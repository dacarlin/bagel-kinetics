```python
import pandas 
import screed
pandas.options.display.max_rows = 1000 
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import sqrt, diag, exp, linspace, log, nan
from scipy.stats import linregress
from scipy.optimize import curve_fit 
from glob import glob 
```

```python
# collect all the data and concatenate into a single dataframe 
#!cut -d, -f1-9 Bagel\ Data\ Reporting/Plate*csv | tr A-Z a-z | sort -ur | tail -n +2 > Bagel\ Data\ Reporting/cleaned_data.csv
#also had to remove a few empty rows and remove spaces from sample names

df = pandas.read_csv( 'Bagel Data Reporting/cleaned_data.csv' )
orig_len = len( df )
df.columns = [ 'plate', 'mutant', 'a280', 'dilution', 'well', 'maxv', 'kobs', 's', 'correction' ]
df['rate'] = ( df.maxv - df.correction ) * 0.0002 / ( df.a280 * df.dilution * 0.25 / 113330 ) 
df = df[( df['maxv'] > 0 )] # throw away negative rates 
print 'tossing {} observations where rate is negative'.format( orig_len - len( df ) )
print '{} observations from {} plates'.format( len( df ), len( df.plate.unique() ) )

plt.scatter( df.maxv * 100 , df.rate * 100  )
```

```python
def v( x, kcat, km ): 
    '''Michaelis-Menten equation, where x: substrate concentration, kcat: kcat, km: km'''
    return kcat * x / ( km + x )

def si( x, kcat, km, ki ): 
    '''Michaelis-Menten equation with substrate inhibition term, via DOI 10.1002/bies.200900167'''
    return kcat * x / ( km + x + ( x ** 2 / ki ) )

def all_in_one( df ):
    for i, ( a280, df ) in enumerate( df.groupby( 'a280' ) ):        
        try:
            p0 = ( df.rate.max(), df.s.mean() )
            popt, pcov = curve_fit( v, df.s, df.rate, p0=p0 )
            perr = sqrt( diag( pcov ) ) / popt * 100 

            if perr[0] < 50 and popt[0] > 1:
                kcat = popt[0]             
            else:
                kcat = perr[0] = nan 
            if perr[1] < 50 and 0.0005 < popt[1] < 0.08: 
                km = popt[1]
            else:
                km = perr[1] = nan
               
            format_string = { 'kcat': '{:.1f}'.format( kcat ),
                              'kcat_std_err': '{:.1f}'.format( perr[0] if perr[0] < 50 else nan ),
                              'km': '{:0.2f}'.format( km * 1000 ),  
                              'km_std_err': '{:0.1f}'.format( perr[1] if perr[1] < 50 else nan ),
                              'a280': a280 }  

            name = df.mutant.unique() 
            if kcat > 0 and km > 0:
                plt.scatter( df.s, df.rate, alpha=0.5, color='navy' )
                xvalues = linspace( 0, 0.075, 99 )
                plt.plot( xvalues, v( xvalues, kcat, km ), alpha=0.7, color='silver' )
                plt.title( name )
                plt.show()
            else:
                print 'No Michaelis-Menten plot for {}'.format( name )
                plt.scatter( df.s, df.rate, alpha=0.5, color='blue' )
                plt.title( name )
                plt.ylim( -.2, 1.2 )
                plt.text( 'No Michaelis-Menten plot for {}'.format( name ) )
                plt.show()

            return pandas.Series( format_string, name='{} ({})'.format( name, i ) ) 
            # {'kcat': '{:.0f}  {:.2f}%'.format( kcat, perr[0] ) , 'km': '{}  {}'.format( km * 1000, perr[1] )  } )
        except Exception as e:
            return e

fits = df.groupby( 'mutant' ).apply( all_in_one )
fits[ [ 'kcat', 'km' ] ] = fits[ [ 'kcat', 'km' ] ].astype( float ) 
fits['kcat_km'] = fits.kcat / fits.km

wild_types = [  ]
for i in fits.index:
    if 'wt' in i or i[0] == i[-1] == 'a':
        wild_types.append( i )

drop = [ i for i in fits.index if '+' in i or '_' in i ]
avg_wt = fits.loc[ wild_types ].mean()
clean_fits = fits.drop( wild_types + drop )
clean_fits
```

```python
# clean up the off-by-threes
for record in screed.open( '/Users/alex/Documents/bagel-orders/wt_fasta/bglb.pep' ):
    seq = record.sequence
    
fit_set = set( clean_fits.index )

print '{} unique samples (including WTs)'.format( len( df['mutant'].unique() ) )
print '{} fits'.format( len( fits ) )

fixer_dict = { }
still_wrong = [ ]

for m in fit_set:
    if len( m ) < 6 and 'wt' not in m:
        if seq[ int( m[1:-1] ) - 1 ].lower() == m[0]: # correct 
            fixer_dict.update( { m: m } )
        elif seq[ int( m[1:-1] ) - 4 ].lower() == m[0]: # off by three
            new_index = '{0}{2}{1}'.format( m[0], m[-1], int( m[1:-1] ) - 3 )
            # print '>>> Correcting: original {} to {}'.format( m, new_index )
            fixer_dict.update( { m: new_index } ) 
        else:
            #print '>>> Error: {}'.format( m )
            still_wrong.append( m )
            
#print new_set
print 'still wrong:', still_wrong
```

```python
def fixer( x ):
    return fixer_dict[ x ]

clean_fits.drop( still_wrong, inplace=True )
clean_fits.index = clean_fits.index.map( fixer )
clean_fits.index = clean_fits.index.map( str.upper )

# add a line with the average WT values 
clean_fits.loc['BglB'] = avg_wt 

# write out to CSV 
clean_fits.to_csv( 'clean_fits.csv' )
```
