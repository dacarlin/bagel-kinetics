```python
import pandas 
import screed
pandas.options.display.max_rows = 1000 
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import sqrt, diag, exp, linspace, log, nan
from scipy.optimize import curve_fit 
from glob import glob 
```

```python
# collect all the data and concatenate into a single dataframe 
#!cut -d, -f1-9 Bagel\ Data\ Reporting/Plate*csv | tr A-Z a-z | sort -ur | tail -n +2 > Bagel\ Data\ Reporting/cleaned_data.csv
#also had to remove a few empty rows and remove spaces from sample names

df = pandas.read_csv( 'Bagel Data Reporting/cleaned_data.csv' )
orig_len = len( df )
df.columns = [ 'plate', 'mutant', 'a280', 'dilution', 'well', 'maxv', 'kobs', 's', 'correction' ]
df['rate'] = ( df.maxv - df.correction ) * 0.002 / ( df.a280 * df.dilution * 0.25 / 113330 ) 
df = df[( df['maxv'] > 0 )] # throw away negative rates 
print 'tossing {} observations where rate is negative'.format( orig_len - len( df ) )
print '{} observations from {} plates'.format( len( df ), len( df.plate.unique() ) )
```

```python
def v( x, kcat, km ): return kcat * x / ( km + x )

def all_in_one( df ):
    
    df['rate']
    # try the MM fit 
    try:
        p0 = ( df.rate.max(), df.s.mean() )
        popt, pcov = curve_fit( v, df.s, df.rate, p0=p0 )
        perr = sqrt( diag( pcov ) )
        km = popt[1] if popt[1] > perr[1] and 0.0005 < popt[1] < 0.07 else nan
        kcat = popt[0] if popt[0] > perr[0] else nan
        #kcat = 10 if popt[0] < 10 else popt[0]
        return pandas.Series( {'kcat': kcat, 'km': km} )
    except Exception as e:
        return e

fits = df.groupby( 'mutant' ).apply( all_in_one )
fits['kcat_km'] = fits.kcat/fits.km
fits.head()
```

```python
wild_types = [  ]
for i in fits.index:
    if 'wt' in i or i[0] == i[-1] == 'a':
        wild_types.append( i )

drop = [ i for i in fits.index if '+' in i or '_' in i ]
avg_wt = fits.loc[ wild_types ].mean()
clean_fits = fits.drop( wild_types + drop + [ 's16a' ] )
clean_fits
```
